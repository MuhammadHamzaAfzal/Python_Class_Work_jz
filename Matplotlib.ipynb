```
def walk():
  x = np.random.randn(50)
  return np.cumsum(x)



X = walk()  
Y = walk()  
Z = walk()  


plt.plot(X)
plt.plot(Y)
plt.plot(Z)

plt.xlabel("Xlabel")
plt.ylabel("Ylabel")
plt.title("Title")
```

```
plt.plot(X)
plt.plot(Y)
plt.plot(Z)

plt.title("Random Walk")
plt.suptitle("Graph", fontsize = 16 , fontweight = "bold")


plt.xlabel("My WALK ON MONDAY")
plt.ylabel("My WALK ON TUESDAY")
```

```
plt.plot(X , label = "X")
plt.plot(Y , label = "Y")
plt.plot(Z , label = "Z")


plt.legend(loc = "lower left")

plt.title("Random Walk" , fontsize = 10)
plt.suptitle("Graph", fontsize = 16 , fontweight = "bold")


plt.xlabel("My WALK ON MONDAY")
plt.ylabel("My WALK ON TUESDAY")
```
