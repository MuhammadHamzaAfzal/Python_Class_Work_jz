```
def walk():
  x = np.random.randn(50)
  return np.cumsum(x)



X = walk()  
Y = walk()  
Z = walk()  


plt.plot(X)
plt.plot(Y)
plt.plot(Z)

plt.xlabel("Xlabel")
plt.ylabel("Ylabel")
plt.title("Title")
```

```
plt.plot(X)
plt.plot(Y)
plt.plot(Z)

plt.title("Random Walk")
plt.suptitle("Graph", fontsize = 16 , fontweight = "bold")


plt.xlabel("My WALK ON MONDAY")
plt.ylabel("My WALK ON TUESDAY")
```

```
plt.plot(X , label = "X")
plt.plot(Y , label = "Y")
plt.plot(Z , label = "Z")


plt.legend(loc = "lower left")

plt.title("Random Walk" , fontsize = 10)
plt.suptitle("Graph", fontsize = 16 , fontweight = "bold")


plt.xlabel("My WALK ON MONDAY")
plt.ylabel("My WALK ON TUESDAY")
```

# Histogram charts

```
plt.hist(y,25)
plt.show()
```

```
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```
objects = ('Python', 'Java Script', 'R', 'Ruby', 'Scala', 'Java')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

```
objects = ('Python', 'R', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.barh(y_pos, performance, align='center', alpha=1)
plt.yticks(y_pos, objects)
plt.xlabel('Usage')
plt.title('Programming language usage')

plt.show()
```
